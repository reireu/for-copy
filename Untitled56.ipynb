```python
import numpy as np
import pandas as pd
import japanize_matplotlib
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# データの準備
df = pd.DataFrame({
    'ad_cost': [10, 20, 30, 40, 50],  # 広告費（万円）
    'sales':   [40, 50, 65, 70, 85]   # 売上（万円）
})

X = df[['ad_cost']]
y = df['sales']

# モデルの学習
model = LinearRegression()
model.fit(X, y)

# 回帰係数と切片
a = model.coef_[0]
b = model.intercept_
r2 = model.score(X, y)

print(f"回帰式: y = {a:.2f} * x + {b:.2f}")
print(f"決定係数 R^2: {r2:.3f}")

# 広告費35万円のときの売上予測
pred = model.predict([[35]])
print(f"広告費35万円のときの予測売上: {pred[0]:.2f} 万円")

# グラフ表示
x_range = np.linspace(0, 60, 100)
y_range = model.predict(x_range.reshape(-1, 1))

plt.scatter(df['ad_cost'], df['sales'], label='実測値')
plt.plot(x_range, y_range, color='red', label='回帰直線')
plt.xlabel('広告費（万円）')
plt.ylabel('売上（万円）')
plt.title('広告費と売上の回帰分析')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

# 予測値と残差の計算
y_pred = model.predict(X)
residuals = y - y_pred

# 残差プロット
plt.figure(figsize=(8, 4))
plt.scatter(df['ad_cost'], residuals, color='blue')
plt.axhline(0, color='red', linestyle='--')
plt.xlabel('広告費（万円）')
plt.ylabel('残差（実測値 - 予測値）')
plt.title('残差プロット（単回帰）')
plt.grid(True)
plt.show()

# 評価指標の計算
mae = mean_absolute_error(y, y_pred)
mse = mean_squared_error(y, y_pred)
rmse = np.sqrt(mse)

print(f"MAE: {mae:.2f}")
print(f"MSE: {mse:.2f}")
print(f"RMSE: {rmse:.2f}")
```

```python
# 重回帰用データ
df_multi = pd.DataFrame({
    'ad_cost': [10, 20, 30, 40, 50],
    'duration': [5, 10, 10, 15, 20],
    'media_type': [0, 1, 0, 1, 1],    # （0: チラシ, 1: SNS）
    'sales': [40, 50, 65, 70, 85]
})

X_multi = df_multi[['ad_cost', 'duration', 'media_type']]
y_multi = df_multi['sales']

model_multi = LinearRegression()
model_multi.fit(X_multi, y_multi)

# 回帰係数と切片
coefs = model_multi.coef_
intercept = model_multi.intercept_
r2_multi = model_multi.score(X_multi, y_multi)

print(f"重回帰式: y = {coefs[0]:.2f}*ad_cost + {coefs[1]:.2f}*duration + {coefs[2]:.2f}*media_type + {intercept:.2f}")
print(f"重回帰モデルの決定係数 R^2: {r2_multi:.3f}")
```

```python
print("\n【評価指標の比較】")
print(f"{'指標':<10}{'単回帰':>10}{'重回帰':>10}")
print(f"{'R^2':<10}{r2:>10.3f}{r2_multi:>10.3f}")
print(f"{'MAE':<10}{mae:>10.2f}{mae_multi:>10.2f}")
print(f"{'MSE':<10}{mse:>10.2f}{mse_multi:>10.2f}")
print(f"{'RMSE':<10}{rmse:>10.2f}{rmse_multi:>10.2f}")
```
