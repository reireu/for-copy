```python
import numpy as np
import pandas as pd
import japanize_matplotlib
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# データの準備
df = pd.DataFrame({
    'ad_cost': [10, 20, 30, 40, 50],  # 広告費（万円）
    'sales':   [40, 50, 65, 70, 85]   # 売上（万円）
})

X = df[['ad_cost']]
y = df['sales']

# モデルの学習
model = LinearRegression()
model.fit(X, y)

# 回帰係数と切片
a = model.coef_[0]
b = model.intercept_
r2 = model.score(X, y)

print(f"回帰式: y = {a:.2f} * x + {b:.2f}")
print(f"決定係数 R^2: {r2:.3f}")

# 広告費35万円のときの売上予測
pred = model.predict([[35]])
print(f"広告費35万円のときの予測売上: {pred[0]:.2f} 万円")

# グラフ表示
x_range = np.linspace(0, 60, 100)
y_range = model.predict(x_range.reshape(-1, 1))

plt.scatter(df['ad_cost'], df['sales'], label='実測値')
plt.plot(x_range, y_range, color='red', label='回帰直線')
plt.xlabel('広告費（万円）')
plt.ylabel('売上（万円）')
plt.title('広告費と売上の回帰分析')
plt.legend()
plt.grid(True)
plt.show()
```
